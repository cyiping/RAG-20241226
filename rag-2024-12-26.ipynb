```python
# 設置 TESSDATA_PREFIX
import os
os.environ["TESSDATA_PREFIX"] = "/usr/share/tesseract-ocr/4.00/"

from pdf2image import convert_from_path
from pytesseract import image_to_string
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.vectorstores import Chroma
from langchain_ollama import OllamaEmbeddings
from langchain.schema import Document  # 引入 Document 類型
import time

# 嵌入模型配置
llm_model = "llama3.2"
embeddings = OllamaEmbeddings(model=llm_model)

# OCR 提取圖片型 PDF 的文字
def extract_text_with_ocr(pdf_path):
    pages = convert_from_path(pdf_path)
    text_pages = []
    start_time = time.time()  # 記錄開始時間
    for i, page in enumerate(pages):
        # 記錄頁處理開始時間
        page_start_time = time.time()

        # 提取文字
        try:
            text = image_to_string(page, lang='chi_tra')  # 'chi_tra' 用於繁體中文
            text_pages.append(Document(page_content=text, metadata={"page": i + 1}))  # 將內容包裝為 Document
            
            # 將內容寫入文件
            file_name = f"page{i + 1}.txt"
            with open(file_name, "w", encoding="utf-8") as file:
                file.write(text)
            
            # 打印頁數提示、提取內容和處理時間
            elapsed_time = time.time() - page_start_time
            print(f"Read No.{i + 1} page")
            print(f"Content saved to {file_name}")
            print(f"Time taken for page {i + 1}: {elapsed_time:.2f} seconds\n")
        except Exception as e:
            print(f"Error processing page {i + 1}: {e}")
    return text_pages

# 創建檢索器
def load_and_retrieve_docs_with_ocr(pdf_path):
    docs = extract_text_with_ocr(pdf_path)  # 使用 OCR 提取文字
    if not docs:
        raise ValueError("OCR extraction failed. Please check the PDF content.")
    
    # 分塊處理
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
    splits = text_splitter.split_documents(docs)
    if not splits:
        raise ValueError("The PDF was not split into any chunks. Please check the extracted text.")
    
    # 向量存儲
    vectorstore = Chroma.from_documents(documents=splits, embedding=embeddings)
    return vectorstore.as_retriever()

# 定義 RAG 鏈條
def rag_chain(question, retriever):
    retrieved_docs = retriever.invoke(question)
    formatted_context = "\n\n".join(doc.page_content for doc in retrieved_docs)
    formatted_prompt = f"Question: {question}\n\nContext: {formatted_context}"
    response = ollama.chat(model=llm_model, messages=[{'role': 'user', 'content': formatted_prompt}])
    return response['message']['content']

# 執行流程
pdf_path = "柯文哲起訴書-2024-12-26.pdf"  # 替換為您的 PDF 檔案路徑
retriever = load_and_retrieve_docs_with_ocr(pdf_path)
result = rag_chain("請總結這份文件？", retriever)

# 輸出結果
from IPython.display import Markdown, display
display(Markdown(result))
```

# 圖片式 pdf 
- 範例 : 柯文哲起訴書

```python
# 設置 TESSDATA_PREFIX
import os
os.environ["TESSDATA_PREFIX"] = "/usr/share/tesseract-ocr/4.00/"

from pdf2image import convert_from_path
from pytesseract import image_to_string
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.schema import Document  # 引入 Document 類型
from transformers import AutoTokenizer, AutoModel
import numpy as np
import faiss
import time
import torch

# 加載 Hugging Face 模型
tokenizer = AutoTokenizer.from_pretrained("sentence-transformers/all-MiniLM-L6-v2")
model = AutoModel.from_pretrained("sentence-transformers/all-MiniLM-L6-v2")

# 嵌入生成函數
def generate_embeddings(texts):
    """使用 Hugging Face Transformers 生成嵌入向量"""
    inputs = tokenizer(
        texts,
        padding=True,
        truncation=True,
        return_tensors="pt"
    )
    with torch.no_grad():
        outputs = model(**inputs)
        embeddings = outputs.pooler_output
    return embeddings.detach().numpy()

# 去除多餘空格函數
def remove_extra_spaces(text):
    """移除 OCR 結果中的多餘空格"""
    return ''.join(text.split())

# OCR 提取或直接讀取已保存的文本
def extract_text_with_ocr_or_load(pdf_path):
    pages = convert_from_path(pdf_path)
    text_pages = []
    start_time = time.time()  # 記錄開始時間
    for i, page in enumerate(pages):
        # 構建文件名
        file_name = f"page{i + 1}.txt"

        if os.path.exists(file_name):  # 如果文件已存在
            print(f"File {file_name} found. Loading content...")
            with open(file_name, "r", encoding="utf-8") as file:
                text = file.read()
            text_pages.append(Document(page_content=text, metadata={"page": i + 1}))
        else:  # 否則執行 OCR
            print(f"File {file_name} not found. Running OCR...")
            try:
                raw_text = image_to_string(page, lang='chi_tra')  # 'chi_tra' 用於繁體中文
                cleaned_text = remove_extra_spaces(raw_text)  # 去除多餘空格
                text_pages.append(Document(page_content=cleaned_text, metadata={"page": i + 1}))
                
                # 將內容寫入文件
                with open(file_name, "w", encoding="utf-8") as file:
                    file.write(cleaned_text)
                
                # 打印頁數提示和處理時間
                elapsed_time = time.time() - start_time
                print(f"Read No.{i + 1} page")
                print(f"Content saved to {file_name}")
                print(f"Time taken for page {i + 1}: {elapsed_time:.2f} seconds\n")
            except Exception as e:
                print(f"Error processing page {i + 1}: {e}")
    return text_pages

# 創建檢索器
def load_and_retrieve_docs_with_ocr_or_load(pdf_path):
    docs = extract_text_with_ocr_or_load(pdf_path)  # 使用 OCR 或加載文件
    if not docs:
        raise ValueError("OCR extraction or file loading failed. Please check the PDF content.")
    
    # 分塊處理
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
    splits = text_splitter.split_documents(docs)
    if not splits:
        raise ValueError("The PDF was not split into any chunks. Please check the extracted text.")
    
    # 生成嵌入並構建 FAISS 索引
    texts = [doc.page_content for doc in splits]
    embeddings = generate_embeddings(texts)
    index = faiss.IndexFlatL2(embeddings.shape[1])  # L2 距離索引
    index.add(embeddings)  # 添加嵌入向量

    return index, splits

# 查詢功能
def query_faiss(index, splits, query, top_k=3):
    query_embedding = generate_embeddings([query]).reshape(1, -1)  # 將查詢嵌入向量化
    distances, indices = index.search(query_embedding, top_k)  # 查詢最近鄰
    results = [splits[idx].page_content for idx in indices[0]]  # 獲取相應的內容
    return results

# 主程式
pdf_path = "柯文哲起訴書-2024-12-26.pdf"  # 替換為您的 PDF 檔案路徑
index, splits = load_and_retrieve_docs_with_ocr_or_load(pdf_path)

# 查詢範例
query = "請總結這份文件？"
results = query_faiss(index, splits, query)

# 輸出查詢結果
from IPython.display import Markdown, display
formatted_results = "\n\n".join(results)
display(Markdown(f"### Query Results\n\n{formatted_results}"))
```

```python
# 查詢範例
query = "起訴那些人？"
results = query_faiss(index, splits, query)

# 輸出查詢結果
from IPython.display import Markdown, display
formatted_results = "\n\n".join(results)
display(Markdown(f"### Query Results\n\n{formatted_results}"))
```

```python
# 設置必要模組
import os
from PyPDF2 import PdfReader  # 用於處理文字型 PDF
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.schema import Document  # 引入 Document 類型
from transformers import AutoTokenizer, AutoModel
import numpy as np
import faiss
import time
import torch

# 加載 Hugging Face 模型
tokenizer = AutoTokenizer.from_pretrained("sentence-transformers/all-MiniLM-L6-v2")
model = AutoModel.from_pretrained("sentence-transformers/all-MiniLM-L6-v2")

# 嵌入生成函數
def generate_embeddings(texts):
    """使用 Hugging Face Transformers 生成嵌入向量"""
    inputs = tokenizer(
        texts,
        padding=True,
        truncation=True,
        return_tensors="pt"
    )
    with torch.no_grad():
        outputs = model(**inputs)
        embeddings = outputs.pooler_output
    return embeddings.detach().numpy()

# 提取文字型 PDF 的內容並保存為文件
def extract_text_from_pdf(pdf_path):
    """從文字型 PDF 提取文字並保存每頁到文件"""
    reader = PdfReader(pdf_path)
    text_pages = []
    for i, page in enumerate(reader.pages):
        try:
            text = page.extract_text()  # 提取文字
            if text:  # 確保提取的文字不為空
                # 保存每頁文字到文件
                file_name = f"wpage{i + 1}.txt"
                with open(file_name, "w", encoding="utf-8") as file:
                    file.write(text)
                print(f"Extracted and saved text from page {i + 1} to {file_name}")
                text_pages.append(Document(page_content=text, metadata={"page": i + 1}))
            else:
                print(f"No text found on page {i + 1}")
        except Exception as e:
            print(f"Error processing page {i + 1}: {e}")
    return text_pages

# 創建檢索器
def load_and_retrieve_docs_from_pdf(pdf_path):
    docs = extract_text_from_pdf(pdf_path)  # 提取 PDF 文字
    if not docs:
        raise ValueError("No text extracted from PDF. Please check the file content.")
    
    # 分塊處理
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
    splits = text_splitter.split_documents(docs)
    if not splits:
        raise ValueError("The PDF was not split into any chunks. Please check the extracted text.")
    
    # 生成嵌入並構建 FAISS 索引
    texts = [doc.page_content for doc in splits]
    embeddings = generate_embeddings(texts)
    index = faiss.IndexFlatL2(embeddings.shape[1])  # L2 距離索引
    index.add(embeddings)  # 添加嵌入向量

    return index, splits

# 查詢功能
def query_faiss(index, splits, query, top_k=3):
    query_embedding = generate_embeddings([query]).reshape(1, -1)  # 將查詢嵌入向量化
    distances, indices = index.search(query_embedding, top_k)  # 查詢最近鄰
    results = [splits[idx].page_content for idx in indices[0]]  # 獲取相應的內容
    return results

# 主程式
pdf_path = "高虹安-李正皓-113年度訴字第27號.pdf"  # 替換為您的 PDF 檔案路徑
index, splits = load_and_retrieve_docs_from_pdf(pdf_path)

# 查詢範例
query = "請總結這份文件內容？"
results = query_faiss(index, splits, query)

# 輸出查詢結果
from IPython.display import Markdown, display
formatted_results = "\n\n".join(results)
display(Markdown(f"### Query Results\n\n{formatted_results}"))
```

```python
# 設置必要模組
import os
from PyPDF2 import PdfReader  # 用於處理文字型 PDF
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.schema import Document  # 引入 Document 類型
from langchain_ollama import OllamaEmbeddings
import numpy as np
import faiss
import time

# 設置 Ollama 模型
llm_model = "llama3.2"
embeddings = OllamaEmbeddings(model=llm_model)

# 提取文字型 PDF 的內容並保存為文件
def extract_text_from_pdf(pdf_path):
    """從文字型 PDF 提取文字並保存每頁到文件"""
    reader = PdfReader(pdf_path)
    text_pages = []
    for i, page in enumerate(reader.pages):
        try:
            text = page.extract_text()  # 提取文字
            if text:  # 確保提取的文字不為空
                # 保存每頁文字到文件
                file_name = f"wpage{i + 1}.txt"
                with open(file_name, "w", encoding="utf-8") as file:
                    file.write(text)
                print(f"Extracted and saved text from page {i + 1} to {file_name}")
                text_pages.append(Document(page_content=text, metadata={"page": i + 1}))
            else:
                print(f"No text found on page {i + 1}")
        except Exception as e:
            print(f"Error processing page {i + 1}: {e}")
    return text_pages

# 創建檢索器
def load_and_retrieve_docs_from_pdf(pdf_path):
    """提取 PDF 文字，生成嵌入並構建 FAISS 索引"""
    docs = extract_text_from_pdf(pdf_path)  # 提取 PDF 文字
    if not docs:
        raise ValueError("No text extracted from PDF. Please check the file content.")
    
    # 分塊處理
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
    splits = text_splitter.split_documents(docs)
    if not splits:
        raise ValueError("The PDF was not split into any chunks. Please check the extracted text.")
    
    # 生成嵌入並構建 FAISS 索引
    texts = [doc.page_content for doc in splits]
    embedding_vectors = np.array(embeddings.embed_documents(texts))  # 使用 Ollama 生成嵌入並轉換為 numpy 數組
    index = faiss.IndexFlatL2(embedding_vectors.shape[1])  # L2 距離索引
    index.add(embedding_vectors)  # 添加嵌入向量

    return index, splits

# 查詢功能
def query_faiss(index, splits, query, top_k=3):
    """基於 FAISS 索引檢索與查詢最相關的文本塊"""
    query_embedding = np.array(embeddings.embed_query(query)).reshape(1, -1)  # 將查詢嵌入向量化並轉換為 numpy
    distances, indices = index.search(query_embedding, top_k)  # 查詢最近鄰
    results = [splits[idx].page_content for idx in indices[0]]  # 獲取相應的內容
    return results

# 主程式
pdf_path = "高虹安-李正皓-113年度訴字第27號.pdf"  # 替換為您的 PDF 檔案路徑
index, splits = load_and_retrieve_docs_from_pdf(pdf_path)

# 查詢範例
query = "請總結這份文件內容？"
results = query_faiss(index, splits, query)

# 輸出查詢結果
from IPython.display import Markdown, display
formatted_results = "\n\n".join(results)
display(Markdown(f"### Query Results\n\n{formatted_results}"))
```
